## Discussion
### What criteria did you use to identify the template with the lowest resolution and highest sequence identity? Why? Did your process differ between SWISS-MODEL and MODELLER? Why or why not?

When the template with the lowest resolution and highest sequence identity is the same template, it means we have found the best template that not only has good structural information, but also is highly similar to the target sequence on a sequence level; therefore, this template is chosen. When they are not the same template, we first filter out all templates with resolutions less than 2A. This is because structures with resolutions lower than 2Å provide detailed atomic position information and are of higher quality. Then, we choose the template with the highest sequence identity to maximize the model's predictive accuracy.

Firstly, SWISS-MODEL is a user-friendly interactive website that allows direct template selection and model building on the website. In contrast, MODELLER uses scripts run on the command line, allowing users to better customize the modeling process by editing the script. Secondly, SWISS-MODEL automatically finds templates with high sequence identity to the target sequence in the database, while MODELLER supports manual template selection. Thirdly, SWISS-MODEL uses QMEAN and Z-score to assess model quality, whereas MODELLER evaluates using the DOPE score and allows manual checking and modification of the model.



### Which of the predicted local similarity to target (local model quality: DOPE, Discrete Optimized Protein Energy) and the overall model quality (Z-score) for the generated model provide the most useful information regarding the generated model? Why?

The DOPE score is based on statistical mechanics principles. It compares the distances of atoms in the model to those in databases protein structures, and measures the model's energy state by calculating the free energy of each amino acid residue in the model. DOPE score is used to assess the quality of local structures in the protein model, helping to identify the state of specific areas in the model.

The Z-score compares the quality of our model to a large number of protein structural models, assessing the overall quality of the model, and helping to evaluate whether the model deviates from the normal range of known protein structures on the whole.

These two parameters reflect different aspects of the model. The DOPE score is more helpful in checking specific areas of the model, while the Z-score evaluates the model's overall reliability and whether it has the correct folding structure. Therefore, when building a model, both parameters should be checked. We can first use the Z-score for an overall assessment, and then the DOPE score for detailed analysis of specific areas.



### Why would the QMEAN provide us with a statistical determination of how good the generated model is? Specifically, think about what contributes to calculation of the QMEAN.

QMEAN uses statistical potentials of mean force to quantify the accuracy of the entire model, as well as modeling errors for each residue and globally. It generates an overall score on how close the model is to known proteins by comparing distances between residue pairs in the model to those in a large database of known structures, distributions of angles in the model to those in known protein structures, and distributions of secondary structures in the model to those in known structures, among other factors. By comparing to structures of measured proteins, QMEAN scores the model, and the closer the score is to the corresponding score of experimentally determined structures, the more likely the model is close to the real protein structure.



### What goes into the calculation of the DOPE score? What are some defining secondary and tertiary structural elements in the structure you selected based on DOPE score?

The DOPE score calculates the energy of a model based on atomic distances and statistical mechanics. It calculates the distance between all pairs of atoms in the protein model, compares them to distances between similar pairs of atoms in databases, and calculates the pairing energy based on the position and type of residue pairs. Additionally, it also considers effects such as solvent. By calculating the DOPE score, we can identify the prediction situation of specific areas in the model. A low DOPE score indicates that specific areas of the model are stable, while a high DOPE score might suggest that the area is unstable or poorly predicted.

The two β-pleated sheets are located on both sides, each surrounded by 4 α-helices. ### What do the RMSD, Z-score, Identity (%), and Similarity (%) tell you about the agreement between the two structures? Specifically, consider each individual contribution to your understanding of structural agreement.


RMSD reflects the degree to which atoms in various parts deviate from average positions, evaluating the consistency between the predictive model structure and experimental structure. A low RMSD value indicates that the two structures are very close, with a high degree of atomic position match; a high RMSD value indicates significant structural differences.

Z-score represents the overall model quality, measuring the structure's total energy deviation from the energy distribution. It is used to evaluate the consistency between the predictive model structure and experimental structure. A low Z-score indicates the model's quality is similar to the experimental structure, while a high Z-score indicates the model is abnormal compared to known structures.

Identity refers to the percentage of identical residues in the total length of matched sequence alignment between two sequences. A high Identity indicates that the two protein sequences are highly conserved, possibly having similar structures and functions. 

Similarity refers to the percentage of identical and similar residues in the total number of comparisons. Compared to identity, Similarity considers the physicochemical properties of amino acids, assessing the similarity in structure and function. This is because two sequences, even if not identical, similar amino acids can have similar properties, bringing similar structures and functions.
